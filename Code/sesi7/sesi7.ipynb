mari lihat jumlah imigran dari jepang (baris 87 ) untuk skenario berikut ini:

1. data baris lengkap (semua kolom)
2. untuk tahun 2013
3. untuk tahun 1980 hingga 1985

```python
# 1. thhe full row data (all columns)
print(df_can.loc['Japan'])

# alternate methods
print(df_can.iloc[87])
print(df_can[df_can.index == 'Japan'].T.squeeze)
```

```python
# 2. for yeaer 2013
print(df_can.loc['Japan', 2013])

#  alternate method
print(df_can.iloc[87, 36])
# year 2013 is the last column, with a positional index of 36
```

```python
#  3. fto years 1980 to 1985
print(df_can.loc['Japan', [1980, 1981, 1982, 1983, 1984, 1985]])
print(df_can.iloc[87, [3, 4, 5, 6, 7, 8]])
```

## Area Plot

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)

df_top5 = df_can.head()

df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(
    kind='area'
    ,stacked=False
    ,figsize=(20,10)
)

plt.title('Immigration Trend of top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(
    kind='area'
    ,alpha=0.25
    ,stacked=False
    ,figsize=(20,10)
)

plt.title('Immigration Trend of top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

## Two types of plotting

```python
# Opsi1
df_top5.plot(
    kind='area'
    ,alpha=0.35
    ,figsize=(20, 10)
)
plt.title('Immigration Trend of top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```

```python
# Opsi2
ax = df_top5.plot(
    kind='area'
    ,alpha=0.35
    ,figsize=(20, 10)
)
ax.set_title('Immigration Trend of top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

```python
df_least5 = df_can.tail(5)
df_least5 = df_least5[years].transpose()

df_least5.plot(
    kind='area'
    ,alpha=0.45
    ,stacked=True
    ,figsize=(20, 10)
)

plt.title('Immigration Trend of top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
ax = df_least5.plot(
    kind='area'
    ,alpha=0.45
    ,stacked=True
    ,figsize=(20, 10)
)

ax.set_title('Immigration Trend of top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

### Pie Charts
Step 1: Gather data

```python
df_continents = df_can.groupby('Continent', axis=0).sum()

print(type(df_can.groupby('Continent', axis=0)))

df_continents.head()
```

Step 2: plot datanya, dengan menambahkan kind='pie'

```python
df_continents['Total'].plot(
    kind='pie',
    figsize=(5,6),
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
)
plt.title('Immigration to Canada by Continent [1980-2013]')
plt.axis('equal')

plt.show()
```

Percantik Pie chart

```python
color_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.05, 0.15]

df_continents['2013'].plot(
    kind='pie',
    figsize=(15,6),
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    labels=None,
    pctdistance=1.12,
    colors=color_list,
    explode=explode_list,
)

plt.title('Immigration to Canada by Continent [1980-2013]', y=1.12)
plt.axis('equal')

# plt.legend(labels=df_continents.index, loc='upper left')
plt.legend(labels=df_continents.index, loc='upper right')
plt.show()
```

### Scatter Plot

```python
df_tot = pd.DataFrame(df_can[years].sum(axis=0))
df_tot.index =map(int, df_tot.index)
df_tot.reset_index(inplace=True)

df_tot.columns = ['year', 'total']

df_tot.head()
```

```python
df_tot.plot(
    kind='scatter',
    x='year',
    y='total',
    figsize=(10,6),
    color='darkblue'
)
plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')

plt.show()
```

```python
df_countries = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_total = pd.DataFrame(df_countries.sum(axis=1))
df_total.reset_index(inplace=True)
df_total.columns= ['year', 'total']
df_total['year']= df_total['year'].astype(int)
df_total.head()
```

```python
df_total.plot(
    kind='scatter',
    x='year',
    y='total',
    figsize=(10,6),
    color='darkblue'
)
plt.title('Immigration from Denmark, Norway, and Sweden to Canada from 1980 -2013')
plt.xlabel('Year')
plt.ylabel('Nuimber of Immigrants')
plt.show()
```
