## sort by "Total

```python

canada_immigration.sort_values(by='Total', ascending=False, axis=0, inplace= True)
top5 = canada_immigration.head(5)
top5 = top5[years].transpose()

top5.plot(kind='line', figsize=(14, 8))
plt.title('immigrants from india, china and phillippines')
plt.xlabel('Year')
plt.ylabel("Number of immigrants")
```

1. top 5 = ascending=false
2. axis=0 urutin berdasarkan row
3. inplace= true = setelah diubah disimpen di canada_immigration
4. figsize = untuk membuat ukuran canvas dalam ukuran inch

### area plot
coba kita tampilin secara area, kind nya diubah jadi area dan tambahin stacked=False

```python
canada_immigration.sort_values(by='Total', ascending=False, axis=0, inplace= True)
top5 = canada_immigration.head(5)
top5 = top5[years].transpose()

top5.plot(kind='area',
          alpha=0.3 ,
          stacked=False ,
          figsize=(14, 8))
plt.title('immigrants from india, china and phillippines')
plt.xlabel('Year')
plt.ylabel("Number of immigrants")
```

### PIE CHART

```python
continent_data = canada_immigration.groupby("Continent", axis=0).sum()
```

1. groupby mengambil data tengah

```python
continent_data
```

```python
continent_data['Total'].plot(
    kind='pie'
    , autopct='%1.1f%%'
    , startangle=45
    , shadow=True
)
```

```python
color_list = ["gold", "yellowgreen", "lightcoral", "lightskyblue", "lightgreen", "pink"]

continent_data['Total'].plot(
    kind='pie'
    ,figsize=(15, 6)
    , autopct='%1.1f%%'
    , startangle=90
    # , shadow=True
    , labels=None
    , colors= color_list
    , pctdistance=1.12
    , explode =[0.1,0,0,0,.1,.1]
)
plt.title("Immigration to Canada by Continent")
plt.axis('equal')
plt.legend(labels=continent_data.index, loc='upper left')
```

1. pctdistance = jarak untuk teks (persen)
= 1.12 >> 1 itu radius / pinggir mentok dan 12 itu 12 persen nya dari radius
2. explode = jarak dari center pie

### SUB PLOT

```python
fig = plt.figure()

ax0 = fig.add_subplot(1,2,1)
ax1 = fig.add_subplot(1,2,2)
# ax2 = fig.add_subplot(2,2,3)
# ax3 = fig.add_subplot(2,2,4)


china_india.plot(kind="box", figsize=(20,6), vert=False, ax=ax0)
ax0.set_title("Boxplot China vs India")
ax0.set_xlabel("Number of Immigrants")
ax0.set_ylabel("Countries")

china_india.plot(kind='line', figsize=(20,6), ax=ax1)
ax1.set_title("Line China vs India")
ax1.set_xlabel("Number of Immigrants")
ax1.set_ylabel("Countries")
```
