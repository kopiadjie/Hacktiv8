```python
df_incidents = pd.read_csv('Police_Department_Incidents_-_Previous_Year__2016_.csv')
print('data ready')
```

```python
df_incidents.head()
```

```python
df_incidents.shape
```

100 insiden pertama dalam kumpulan data ini

```python
limit = 100
df_incidents = df_incidents.iloc[0:limit, :]
```

```python
df_incidents.shape
```

```python
latitude = 37.77
longitude = -122.42
```

```python
sanfran_map = folium.Map(location=[latitude,longitude], zoom_start=12)
sanfran_map
```

```python
incidents = folium.map.FeatureGroup()

for lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='yellow',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6
        )
    )
sanfran_map.add_child(incidents)
```

Menambahkan teks pop jika diklik

```python
incidents = folium.map.FeatureGroup()

for lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='yellow',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6
        )
    )

# add pop text
latitudes = list(df_incidents.Y)
longitude = list(df_incidents.X)
labels = list(df_incidents.Category)

for lat,lng, label in zip(latitudes, longitude, labels):
    folium.Marker([lat,lng], popup=label).add_to(sanfran_map)

sanfran_map.add_child(incidents)
```

```python
sanfran_map = folium.Map(location=[latitude,longitude], zoom_start=12)

for lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='yellow',
        fill=True,
        popup=label,
        fill_color='blue',
        fill_opacity=0.6,
        ).add_to(sanfran_map)
    
sanfran_map
```
